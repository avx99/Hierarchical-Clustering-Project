# Exemple Simple Classification Hiérarchique

## Importing the libraries
- Pandas : Pour gerer et manipuler les structures de données multidimentionnels.
- Numpy : Pour gerer les calculs matriciels.
- matplotlib.pyplot : pour visualiser les données.
- seaborn : pour la visualisation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the dataset

```python
#read_csv : est une fonction pour l'importation des données (csv,tsv,txt ...) .
#le premier parameétre c'est l'url vers notre dataset ,le deuxiéme paramétre c'est le sparateur ,dans notre cas c'est '\t' qui
#est une tabulation car notre fichier est de type tsv
fromages = pd.read_csv('http://www.math.u-bordeaux.fr/~mchave100p/wordpress/wp-content/uploads/2013/10/fromage.txt', sep='\t') 
```

```python
#visualisation de notre dataset
fromages
```

```python
#en met ici le variable "fromages" comme index car les élements de "fromages" sont unique.
#inplace=True sert a remplacer l'ancien dataset par la nouvelle(aui contient "fromages" comme index) sur place
#si inplace=false => il faut faire : fromages = fromages.set_index('Fromages', inplace=False).inplace=False par défault.
fromages.set_index('Fromages', inplace=True)
#visualisation de notre dataset
fromages
```

## Matrice des données

```python
#en utilise iloc.values pour rendre X comme un tableau numpy pour faciliter l'étude.
X = fromages.iloc[:,:].values
#Dataframe sert a rendre X de type pandas pour bien visualiser X.
pd.DataFrame(X)
```

```python
# Nombres des individus
n = len(X)

# Nombres des variables
p = len(X[0])
```

## Les moyennes des variable

```python
#mean sert a calculer les moyennes Xj de X,sachant que Xj c'est la moyenne de la colone j de X
#axis = 0 pour préciser que on va appliquer la moyenne sur les colones.si on veut calculer les moyemme Xi des individus il faut utiliser axis = 1.
mean = X.mean(axis=0)
#{'Moyenne':mean} sert a nommer mean par Moyenne
#T c'est le le transposé de mean
pd.DataFrame({'Moyenne':mean}).T
```

## La variance pour chaque variable

```python
#mean sert a calculer les Ecarts types Xj de X,sachant que Xj c'est l'Ecart type de la colone j de X
#axis = 0 pour préciser que on va appliquer l'Ecart type sur les colones
std = X.std(axis=0)
#'Ecart type':std sert a nommer std par Ecart type
#T c'est le le transposé de std
pd.DataFrame({'Ecart type':std}).T
```

## Matrice des données centré reduits

```python
#la normalisation
Z = (X-mean)/std
#visualisation de la matrice Z centée-reduite.
pd.DataFrame(Z)
```

```python
#StandardScaler est une classe qui peut nous aider a norlaliser notre matrice X.
from sklearn.preprocessing import StandardScaler
#sc est une instance de StandardScaler.
sc = StandardScaler()
#fit_transform est une méthode de la classe StandardScaler qui prend en paramétre la matrice X et il retourne la matrice Z
#en peut séparer la méthode fit_transform en duex méthode :
# 1- la méthode fit(X) : sert a donner X a l'objet sc. cette méthode n'a pas une valeur a retourner
# 2- la méthode transfrom() : sert a retourner le Z
#alors on a Z2 = sc.fit_transform(X) == sc.fit(X);Z2 = sc.transform()
Z2 = sc.fit_transform(X)
pd.DataFrame(Z2)
```

## Application de ACP

```python
#pour appliquer l'ACP (PCA : Principal components analysis)
from sklearn.decomposition import PCA
#acp est une instance de la classe PCA.
acp = PCA()
```

### Matrice F des composants principales

```python
#F est une matrice qui contient les composants Fi
F = acp.fit_transform(Z)
#La visualisation de nos composants
#columns=["f"+str(i) for i in range(p)] sert nommer chaque colonne par fi avec 0<=i<=8
pd.DataFrame(F, columns=["f"+str(i) for i in range(p)])
```

### Les valeurs propres

```python
#le calule des valeurs propres , axis = 0 indique qu'on va calculer le std sur les colonnes
valeur_propres = np.std(F,axis=0)
#visualisation du transposé de valeur_propres
pd.DataFrame({'Valeurs Propres':valeur_propres}).T
```

### Pourcentage des valeurs propres

$$ \frac{\lambda_{\alpha}}{\lambda_{1}+....+ \lambda_{p}} $$

```python
#somme contient la somme des élements de valeur_propres
somme = sum(valeur_propres)
#le vecteur pourcentage contient le pourcentage de chaque valeur propre
pourcentage = [valeur_propres[i]*100/somme for i in range(p)]
#visualisation de pourcentage
print(pourcentage)
```

### Visualisation de pourcentage des valeurs propres

```python
#plt.bar sert a créer un histogramme des valeurs propres avec ces pourcentages
# le x contient les élements de 0 a p , y contient les pourcentages
plt.bar(x=range(p) ,height=pourcentage)
```

```python
#visualiser la somme des deux premiers pourcentage des deux premiers axes donnée par l'ACP
#on va conserver 52% d'information si on utilise deux axes
pourcentage[0] + pourcentage[1]
```

## Visualisation de données avec ACP

```python
#dim c'est le dimension de notre graphe , dans ce cas en utilise (10,10) c'est a dire que -5<x<5 et -5<y<5
dim = 10
#initialisation d'un graphe vide de dimension (10,10)
fig, ax = plt.subplots(figsize=(dim,dim))
#utiliser F0 (F[:,0]) comme x et F1 (F[:,1]) comme y
#le deuxième paramétre "o" c'est pour visualiser chaque individu comme "o"
ax.plot(F[:,0],F[:,1],"o")
#-5<x<5 et -5<y<5
ax.axis([-dim/2,+dim/2,-dim/2,+dim/2])
#dessiner le repére en gris de centre (0,0)

#création d'une ligne horizontale dont les cordonnées sont (-5,0) (5,0)
#color='silver' pour utiliser la couleur grise
#linestyle='--' pour une ligne Ligne pointillée , vous pouvez utiliser '-' pour une ligne continue.
ax.plot([-dim/2,+dim/2],[0,0],color='silver',linestyle='--')
#création d'une ligne verticale dont les cordonnées sont (0,-5) (0,5)
#color='silver' pour utiliser la couleur grise
#linestyle='--' pour une ligne Ligne pointillée , vous pouvez utiliser '-' pour une ligne continue.
ax.plot([0,0],[-dim/2,+dim/2],color='silver',linestyle='--')
#tracer le pourcentage de la premiér axe sur x
ax.set_xlabel("Comp.1  "+ str(pourcentage[0])+" %")
#tracer le pourcentage de la deuxième axe sur y
ax.set_ylabel("Comp.2  "+ str(pourcentage[1])+" %")
#pour chaque individu en donne ces cordonnées (F[i,0],F[i,1]) et la troisième paramétre i c'est le nom de cet individu
#dans ce cas en utilise l'indice i comme un nom {0,1,2,3,4...}
for i in range(n):
    ax.text(F[i,0],F[i,1],i)
#plt.show() sert a visualiser notre graphe
plt.show()
```

## Classification

### Nombre de groups
* determinons d abord nombre de groups avec la methode de dendrogram on utilisons la distance <i>Ward</i>

$$ d(u,v) = \sqrt{\frac{|v|+|s|}
                           {T}d(v,s)^2
                    + \frac{|v|+|t|}
                           {T}d(v,t)^2
                    - \frac{|v|}
                           {T}d(s,t)^2}$$

$$ d(A,B) = \frac{n_1*n_2}{n_1+n_2}\sum_{w \in A \cup B} d^2(w,G_{AB})$$

```python
#l'importaion de dendogramme pour la visualisation et linkage pour la classification
from scipy.cluster.hierarchy import dendrogram, linkage
#la création du dendogramme en utilisant la distance de ward
dendogram = dendrogram(linkage(Z,method = 'ward'))
#le titre du dendogramme
plt.title('dendogramme')
#le titre de x
plt.xlabel('les individus')
#le titre de y
plt.ylabel('la distance')
#visualiser le dendogramme
plt.show()
```

```python
#en remarque que le meilleur nombre des classes est 2
n_clusters=2
```

## Training l'algorithme de Classification Hiérarchique  sur les données

```python
#AgglomerativeClustering utiliser pour prédir les resultats de la classification
from sklearn.cluster import AgglomerativeClustering
#hc est une instance de la classe AgglomerativeClustering avec deux classe , en utilisant ward et la distance euclidian
hc = AgglomerativeClustering(n_clusters,affinity='euclidean',linkage='ward')
```

### Prediction des groupe

```python
#y_hc contient les resultats de la prédiction
y_hc = hc.fit_predict(F)
#visualisation de y_hc
pd.DataFrame(y_hc)
```

### Visualisation du groups 2D

```python
#dim c'est le dimension de notre graphe , dans ce cas en utilise (10,10) c'est a dire que -5<x<5 et -5<y<5
dim = 10
#initialisation d'un graphe vide de dimension (10,10)
fig, ax = plt.subplots(figsize=(dim,dim))
#présenter chaque individu avec ces cordonnes en respectant sa classe
#s = 50 c'est le volume de chaque cercle dans le graphe
#c = couleur de cette cercle
#label c'est le label de chaque couleur
plt.scatter(F[y_hc == 0, 0], F[y_hc == 0, 1], s = 50, c = 'red', label = 'Groupe 1')
plt.scatter(F[y_hc == 1, 0], F[y_hc == 1, 1], s = 50, c = 'blue', label = 'Groupe 2')
#-5<x<5 et -5<y<5
ax.axis([-dim/2,+dim/2,-dim/2,+dim/2])
#création d'une ligne horizontale dont les cordonnées sont (-5,0) (5,0)
#color='silver' pour utiliser la couleur grise
#linestyle='--' pour une ligne Ligne pointillée , vous pouvez utiliser '-' pour une ligne continue.
ax.plot([-dim/2,+dim/2],[0,0],color='silver',linestyle='--')
#création d'une ligne verticale dont les cordonnées sont (0,-5) (0,5)
#color='silver' pour utiliser la couleur grise
#linestyle='--' pour une ligne Ligne pointillée , vous pouvez utiliser '-' pour une ligne continue.
ax.plot([0,0],[-dim/2,+dim/2],color='silver',linestyle='--')
#tracer le pourcentage de la premiér axe sur x
ax.set_xlabel("Comp.1  "+ str(pourcentage[0])+" %")
#tracer le pourcentage de la deuxième axe sur y
ax.set_ylabel("Comp.2  "+ str(pourcentage[1])+" %")
#pour visualiser les labels dans le graph
plt.legend()
#visualiser le graph
plt.show()
```
