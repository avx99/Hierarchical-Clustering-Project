## Importing the libraries
- Pandas : Pour gerer et manipuler les structures de données multidimentionnels.
- Numpy : Pour gerer les calculs matriciels.
- matplotlib.pyplot : pour visualiser les données.
- seaborn : pour la visualisation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing the dataset

```python
#read_csv : est une fonction pour l'importation des données (csv,tsv,txt ...) 
dataset = pd.read_csv("default of credit card clients.csv")
```

```python
dataset
```

```python
#on éliminue le 'ID'
dataset= dataset.iloc[:500,1:]
```

### Matrice de données

```python
#en utilise iloc.values pour rendre X comme un tableau numpy pour faciliter l'étude.
X = dataset.iloc[:,:-1].values
# print(X)
pd.DataFrame(X)
```

### Target

```python
#le vrai resultat
y = dataset.iloc[:,-1].values
#Dataframe sert a rendre X de type pandas pour bien visualiser y.
pd.DataFrame(y)
```

## Analyse multivariable

### Pair Plot

```python
#cette representation permet de visualiser tous les graphes possible en fonction de tous les variables.
#on a choisit le style blue et blanc
sns.set_style("whitegrid");
#on a creé le graph
sns.pairplot(dataset[:500], size=3);
#visualisation du graph
plt.show()
```

```python
#le meme travail ms just pour les 50 premiers individu
sns.set_style("whitegrid");
sns.pairplot(dataset[:50]);
plt.show()
```

### Heat map

```python
#heatmap utliser pour visualiser la corrélation entre les variables.
#matrix contient le trainagle inferieur de la matrice de corrélation sans diagonale puisque la diagonale contient des uns et la mtrice est symétrique
matrix = np.triu(dataset.corr())
#creation du graph avec une orientation horizontale
sns.heatmap(dataset.corr(), cbar_kws= {'orientation': 'horizontal'},mask=matrix)
```

## Analyse des composantes principales

### Normalise les données

```python
#StandardScaler est une classe qui peut nous aider a norlaliser notre matrice X.
from sklearn.preprocessing import StandardScaler 
#fit_transform est une méthode de la classe StandardScaler qui prend en paramétre la matrice X et il retourne la matrice Z
#en peut séparer la méthode fit_transform en duex méthode :
# 1- la méthode fit(X) : sert a donner X a l'objet du StandardScaler. cette méthode n'a pas une valeur a retourner
# 2- la méthode transfrom() : sert a retourner le Z
#alors on a Z2 = StandardScaler().fit_transform(X) == StandardScaler().fit(X);Z2 = StandardScaler().transform()
Z = StandardScaler().fit_transform(X)
#visualiser Z
pd.DataFrame(Z)
```

```python
#mean sert a calculer les moyennes Xj de X,sachant que Xj c'est la moyenne de la colone j de X
#axis = 0 pour préciser que on va appliquer la moyenne sur les colones.si on veut calculer les moyemme Xi des individus il faut utiliser axis = 1.
moyennes = np.mean(Z,axis=0)
#T c'est le le transposé de mean
#on apllique sur chaque element des moyennes la partient entier et la valeur absolu car ces element sont tres proch de 0
pd.DataFrame(moyennes).T.apply(lambda x: round(abs(x)))
```

```python
#mean sert a calculer les Ecarts types Xj de X,sachant que Xj c'est l'Ecart type de la colone j de X
#axis = 0 pour préciser que on va appliquer l'Ecart type sur les colones
var = np.std(Z,axis=0)
#T c'est le le transposé de std
pd.DataFrame(var).T
```

### PCA

```python
#pour appliquer l'ACP (PCA : Principal components analysis)
from sklearn.decomposition import PCA
#acp est une instance de la classe PCA.
acp = PCA()
```

### Matrice F des projections selon les axes principales

```python
#composants est une matrice qui contient les composants Fi
composants = acp.fit_transform(Z)
# print(composants)
#columns=["f"+str(i) for i in range(p)] sert nommer chaque colonne par fi avec 0<=i<=22
pd.DataFrame(composants, columns=["f"+str(i) for i in range(len(Z[0]))])
```

### les valeurs propres

```python
#le calule des valeurs propres , axis = 0 indique qu'on va calculer le std sur les colonnes
lambdas = np.std(composants,axis=0)
#visualisation du transposé de lambdas
pd.DataFrame({'lambdas':lambdas})
```

```python
#s contient la somme des élements des valeurs propres
s = sum(lambdas)
#n contient le nombre des valeurs propres
n = len(lambdas)
#v contient le pourcentage des valeurs propres
v = [lambdas[i]*100/s for i in range(n)]
print(v)
```

```python
#plt.bar sert a créer un histogramme des valeurs propres avec ces pourcentages
# le x contient les élements de 0 a n , y contient les pourcentages
plt.bar(x=range(n) ,height=v)
```

### Visualisation ACP

```python
#dim c'est le dimension de notre graphe , dans ce cas en utilise (30,30) c'est a dire que -10<x<10 et -10<y<10
#n  est le nombre des individus a pésenter
n = 500
dim = 30
#initialisation d'un graphe vide de dimension (30,30)
fig, ax = plt.subplots(figsize=(dim,dim))
#utiliser F0 (composants[:,0]) comme x et F1 (composants[:,1]) comme y
#le deuxième paramétre "o" c'est pour visualiser chaque individu comme "o"
ax.plot(composants[:,0],composants[:,1],"o")
#-15<x<15 et -15<y<15
ax.axis([-dim/2,+dim/2,-dim/2,+dim/2])
#dessiner le repére en gris de centre (0,0)

#création d'une ligne horizontale dont les cordonnées sont (-15,0) (15,0)
#color='silver' pour utiliser la couleur grise
#linestyle='--' pour une ligne Ligne pointillée , vous pouvez utiliser '-' pour une ligne continue.
ax.plot([-dim/2,+dim/2],[0,0],color='silver',linestyle='--')
#création d'une ligne verticale dont les cordonnées sont (0,-15) (0,15)
#color='silver' pour utiliser la couleur grise
#linestyle='--' pour une ligne Ligne pointillée , vous pouvez utiliser '-' pour une ligne continue.
ax.plot([0,0],[-dim/2,+dim/2],color='silver',linestyle='--')
#tracer le pourcentage de la premiér axe sur x
ax.set_xlabel("Comp.1  "+ str(v[0])+" %")
#tracer le pourcentage de la deuxième axe sur y
ax.set_ylabel("Comp.2  "+ str(v[1])+" %")
#pour chaque individu en donne ces cordonnées (F[i,0],F[i,1]) et la troisième paramétre i c'est le nom de cet individu
#dans ce cas en utilise l'indice i comme un nom {0,1,2,3,4...}
for i in range(n):
    ax.text(composants[i,0],composants[i,1],i)
#plt.show() sert a visualiser notre graphe
plt.show()
```

### Dendrogram

```python
#l'importaion de dendogramme pour la visualisation et linkage pour la classification
from scipy.cluster.hierarchy import dendrogram, linkage
#la création du dendogramme en utilisant la distance de ward
dendogram = dendrogram(linkage(Z[:50],method = 'ward'))
#le titre du dendogramme
plt.title('dendogramme')
#le titre de x
plt.xlabel('les individus')
#le titre de y
plt.ylabel('la distance')
#visualiser le dendogramme
plt.show()
```

```python
#meme chose mais ici pour Z complet
dendogram = dendrogram(linkage(Z,method = 'ward'))
plt.title('dendogramme')
plt.xlabel('les individus')
plt.ylabel('la distance')
plt.show()
```

```python
#meme chose mais ici pour les deux premiers axes d' acp
dendogram = dendrogram(linkage(composants[:,:2],method = 'ward'))
plt.title('dendogramme')
plt.xlabel('les individus')
plt.ylabel('la distance')
plt.show()
```

```python
#meme chose mais ici avec la distance min
dendogram = dendrogram(linkage(Z,method = 'single'))
plt.title('dendogramme')
plt.xlabel('les individus')
plt.ylabel('la distance')
plt.show()
```

```python
#meme chose mais ici avec la distance max
dendogram = dendrogram(linkage(Z,method = 'complete'))
plt.title('dendogramme')
plt.xlabel('les individus')
plt.ylabel('la distance')
plt.show()
```

```python
#meme chose mais ici avec la distance moyenne
dendogram = dendrogram(linkage(Z,method = 'average'))
plt.title('dendogramme')
plt.xlabel('les individus')
plt.ylabel('la distance')
plt.show()
```

```python
#AgglomerativeClustering utiliser pour prédir les resultats de la classification
from sklearn.cluster import AgglomerativeClustering
#hc est une instance de la classe AgglomerativeClustering avec deux classe , en utilisant ward et la distance euclidian
hc = AgglomerativeClustering(n_clusters=2,affinity='euclidean',linkage='ward')
```

### Classification

```python
#y_hc_acp c'est les classes generent par notre modele en se basant sur les composants acp
y_hc_acp = hc.fit_predict(composants[:,:2])
```

```python
pd.DataFrame(y_hc_acp)
```

```python
#y_hc c'est les classes generent par notre modele en se basant sur Z
y_hc = hc.fit_predict(Z)
```

```python
pd.DataFrame(y_hc)
```

### Comparaison entre les resultats


En apprentissage automatique, la matrice de confusion est une matrice qui mesure la qualité d'un système de classification. Chaque ligne correspond à une classe réelle, chaque colonne correspond à une classe estimée. La cellule ligne L, colonne C contient le nombre d'éléments de la classe réelle L qui ont été estimés comme appartenant à la classe C1.

Un des intérêts de la matrice de confusion est qu'elle montre rapidement si un système de classification parvient à classifier correctement.

```python
from sklearn.metrics import confusion_matrix
cm_acp = confusion_matrix(y_hc_acp, y)
sns.heatmap(cm_acp, annot=True)
cm_acp
```

```python
from sklearn.metrics import confusion_matrix
cm_z = confusion_matrix(y_hc, y)
sns.heatmap(cm_z, annot=True)
cm_z
```
